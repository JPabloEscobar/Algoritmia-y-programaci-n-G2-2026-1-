```python
df = pd.read_csv("Clase Gr2_NOTASMODS (CSV para shofle).csv", sep=";")
df.head()
```

```python
print(df.columns.tolist())
```

```python
df.columns = df.columns.str.strip()
print(df.columns.tolist())
```

```python
Names = df[["Nombres", "EMAIL"]]
print(Names)
```

```python
#Grupos = list(zip(df["Nombres"], df["EMAIL"]))
Grupos = (df["Nombres"] + " (" + df["EMAIL"] + ")").tolist()

for i, pos in enumerate(range(0, len(Grupos), 4), start=1):
    gr = Grupos[pos: pos+4]
    print(f"Grupo{i}: {', '.join(gr)}")
```

```python
import pandas as pd

# 1. Creamos una lista vacía para guardar los datos de los grupos
datos_exportar = []

# 2. Usamos el ciclo que ya conoces
for i, pos in enumerate(range(0, len(Grupos), 4), start=1):
    gr = Grupos[pos : pos+4]
    
    # Unimos los integrantes en un solo texto o los separamos
    # En este caso, los pondremos en una sola celda separados por " y "
    integrantes = " y ".join(gr)
    
    # Guardamos la información en formato de fila (diccionario)
    datos_exportar.append({
        "Grupo": f"Grupo {i}",
        "Integrantes": integrantes
    })

# 3. Convertimos la nueva lista en un DataFrame
df_grupos = pd.DataFrame(datos_exportar)

# 4. Exportamos a Excel
# Nota: Necesitas tener instalada la librería 'openpyxl' 
# (pip install openpyxl)
df_grupos.to_excel("Lista_de_Grupos.xlsx", index=False)

print("¡Archivo 'Lista_de_Grupos.xlsx' creado con éxito!")
```
